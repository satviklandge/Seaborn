## Seaborn Barplot 
    A Seaborn barplot is used to visualize the relationship between a categorical variable and a numerical variable. It displays bars that represent the estimated value (usually the mean) of the numeric data for each category. Bar plots are helpful for comparing values across different groups.

    Seaborn automatically calculates statistical values such as the mean and can also show confidence intervals, making it very useful for quick data analysis and visualization

       -> Key Features

        - Displays categorical vs numerical data
        - Automatically calculates average values
        - Supports error bars / confidence intervals
        - Highly customizable with colors, palettes, and labels
        - Works smoothly with Pandas DataFrames

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
sns.barplot(x = [1,2,3,4], y = [5,6,7,8], )
plt.xlabel('X - Axis ')
plt.ylabel('y - Axis ')
plt.show()
```

### 1 . sns.barplot() x , y parameters

```python
sns.barplot(x = tips_df.day , y = tips_df.total_bill)
```

### 2. sns.barplot() hue parameter

```python
sns.barplot(x = tips_df.day , y = tips_df.total_bill, hue = tips_df.sex)
```

### 3. sns.barplot() data parameter 

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df)
```

### 4. sns.barplot() order parameter 

```python
order = ['Sun' , 'Thur' , 'Fri' , 'Sat']
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , order = order)
```

### 5. sns.barplot() hue_order parameter 

```python
hue_order = ['Female' ,'Male']
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , hue_order = hue_order)
```

###  6. sns.barplot() estimator parameter

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , estimator= np.max)
```

### 7. sns.barplot() ci parameter 

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , ci = 20 )
```

### 8. sns.barplot() n_boot parameter 

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , n_boot=2)
```

### 9. sns.barplot() orient parameter

```python
tips_df.columns
```

```python
sns.barplot( x = 'total_bill' , y = 'size' , hue = 'sex', data = tips_df , orient= 'y' )
```

### 10. sns.barplot() color parameter 

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , color='g' )
```

### 11. sns.barplot() palette parameter

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , palette= "Wistia_r" )
```

### 12. sns.barplot() saturation parameter 

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , saturation=.2 )
```

### 13. sns.barplot() errcolor parameter

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , errcolor='0.7' )
```

### 14. sns.barplot() errwidth parameter 

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , errwidth= 9 )
```

### 15. sns.barplot() capsize parameter 

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , capsize=1)
```

### 16. sns.barplot() dodge parameter 

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , dodge= False)
```

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , dodge= True)
```

### 17. sns.barplot() kwargs parameter 

```python
kwargs = {'alpha':0.9, 'linestyle':':', 'linewidth':5, 'edgecolor':'k'}
```

```python
sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df ,**kwargs)
```

```python
# Pass Keyword argument as parameter 

sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , alpha =.9, linestyle = '-.', linewidth = 3 , edgecolor = 'g')
```

### 18. sns.barplot() ax parameter 

```python
 ax = sns.barplot( x = 'day' , y = 'total_bill' , hue = 'sex', data = tips_df , alpha =.9, linestyle = '-.', linewidth = 3 , edgecolor = 'g')

ax.set( title = "Barplot of Tips DataFrame",
      xlabel = "Days",
      ylabel = "Total Bill")
```
