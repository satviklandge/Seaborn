## Seaborn Histogram
    A Seaborn Histogram is a powerful way to visualize the distribution of numerical data. It shows how your data is spread across different value ranges (called bins), making it easy to understand patterns like skewness, spread, and central tendency.
    
    -> Key Parameters
    
        | Parameter | Description                  |
        | --------- | ---------------------------- |
        | `data`    | Dataset (DataFrame or array) |
        | `x`       | Column name for values       |
        | `bins`    | Number of intervals          |
        | `kde`     | Adds smooth density curve    |
        | `color`   | Bar color                    |
        | `hue`     | Group by category            |


```python
import seaborn as sns
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
bins = [300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500,1600,1700,1800]
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["ins_premium"], bins = bins )
plt.xticks(bins)
plt.title("Histogram of Total Bill")
plt.legend()
```

### 10. sns distplot histogram keyword arguments

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["ins_premium"],
            hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00',
                       'linewidth':5, 'linestyle':'--', 'alpha':0.9})
```

### 11. sns distplot kde(kernel density estimate) keyword arguments 

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["ins_premium"],
            hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00',
                       'linewidth':5, 'linestyle':'--', 'alpha':0.9},
            kde_kws = {'color':'#8e00ce', 'linewidth':8, 'linestyle':'--', 'alpha':0.9})
```

### 12. sns distplot rugplot keyword arguments 

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["ins_premium"],
            hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00',
                       'linewidth':5, 'linestyle':'--', 'alpha':0.9},
            kde_kws = {'color':'#8e00ce', 'linewidth':8, 'linestyle':'--', 'alpha':0.9},
            rug = True,
            rug_kws = {'color':'#0426d0', 'edgecolor':'#00dbff', 'linewidth':3, 'linestyle':'--', 'alpha':0.9},)
```

### 13. sns distplot fit keyword arguments 

```python
plt.figure(figsize=(16,9))
sns.set()
sns.distplot(tips_df["ins_premium"],
            hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00',
                       'linewidth':5, 'linestyle':'--', 'alpha':0.9},
             kde = False,
             fit = norm,
            fit_kws = {'color':'#8e00ce', 'linewidth':12, 'linestyle':'--', 'alpha':0.4},
            rug = True,
            rug_kws = {'color':'#0426d0', 'edgecolor':'#00dbff', 'linewidth':3, 'linestyle':'--', 'alpha':0.9},)
```

### 14. Best way to plot a seaborn histogram 

```python
plt.figure(figsize=(16,9))
sns.set()
bins = [300,400,500,600,700,800,900,1000,1100,1200,1300,1400,1500,1600,1700,1800]
sns.distplot(tips_df["ins_premium"], bins = bins,
                         hist_kws = {'color':'#DC143C', 'edgecolor':'#aaff00',
                       'linewidth':5, 'linestyle':'--', 'alpha':0.9},
             kde = False,
             fit = norm,
            fit_kws = {'color':'#8e00ce', 'linewidth':12, 'linestyle':'--', 'alpha':0.4},
            rug = True,
            rug_kws = {'color':'#0426d0', 'edgecolor':'#00dbff', 'linewidth':3, 'linestyle':'--', 'alpha':0.9},
            label = "TB")
plt.xticks(bins)
plt.title("Histogram of Car Crashes Ins Premium ", fontsize = 20)
plt.xlabel("Ins Premium", fontsize = 15)
plt.legend()
plt.show()
```
