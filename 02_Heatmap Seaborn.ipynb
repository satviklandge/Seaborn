## Seaborn Heatmap 
    Seaborn Heatmap is a data visualization technique used to display values in a matrix format using colors. It helps quickly identify patterns, correlations, and variations in data by representing numbers with different color intensities.
    
    It is part of the Seaborn library, which is built on top of Matplotlib and works well with Pandas data structures.

    🔹 Important Parameters
        | Parameter    | Description                   |
    | ------------ | ----------------------------- |
    | `data`       | Matrix-like dataset           |
    | `annot=True` | Displays numbers inside cells |
    | `cmap`       | Controls color scheme         |
    | `linewidths` | Adds lines between cells      |
    | `fmt`        | Format of annotation values   |


```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
array_2d = np.linspace(1,5,12).reshape(4,3)
print (array_2d)
```

```python
sns.heatmap(array_2d)
```

```python
globalwarming_df = pd.read_csv("Who_is_responsible_for_global_warming.csv")
globalwarming_df.head()
```

```python
globalwarming_df.columns
```

```python
globalwarming_df = globalwarming_df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis=1).set_index('Country Name')
globalwarming_df
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df)
```

### 1. sns.heatmap() vmin & vmax parameter 

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df, vmin = 0, vmax = 21)
```

### 2. How to change seaborn heatmap color using cmap parameter? 

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df, cmap="cividis_r")
```

### 3. Change cmap using sns.heatmap() center parameter 

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df, cmap="cividis_r", center = 10.0)
```

 ### 4. How to remove vmin or vmax using sns.heatmap() robust parameter?  


```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df, robust = True)
```

### 5. Seaborn heatmap annot parameter – add a number on each cell 

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalwarming_df, annot = True)
```

```python
annot_arr = np.arange(1,13).reshape(4,3)
sns.heatmap(array_2d, annot = annot_arr)
```

### 6. sns.heatmap() fmt parameter – add text on each cell 

```python
annot_arr = np.array([['a00', 'a01', 'a02'],
                     ['a10', 'a11', 'a12'],
                      ['a20', 'a21', 'a22'],
                     ['a30', 'a31', 'a32']], dtype = str)
```

```python
sns.heatmap(array_2d, annot = annot_arr, fmt="s") 
```

### How to change style & format of annot (annotate) using sns.heatmap() annot_kws? 

```python
plt.figure(figsize=(16,9))
annot_kws = {'fontsize':10,
            'fontstyle':'italic',
            'color':"k",
            'alpha':0.6,
            'rotation':"vertical",
            'verticalalignment':'center',
            'backgroundcolor':'w'}
sns.heatmap(globalwarming_df, annot = True, annot_kws = annot_kws)
```

### Separate each cell of a heatmap using sns.heatmap() linewidths parameter 

```python
plt.figure(figsize=(16,9)) 
sns.heatmap(globalwarming_df, linewidths=7)
```

### Change the line color of seaborn heatmap with linecolor parameter 

```python
plt.figure(figsize=(16,9)) 
sns.heatmap(globalwarming_df, linewidths=3, linecolor="k")
```

### How to hide color bar using sns.heatmap() cbar parameter? 

```python
plt.figure(figsize=(16,9)) 
sns.heatmap(globalwarming_df, cbar=False)
```

### How to change the color bar of seaborn heatmap? 

```python
plt.figure(figsize=(14,14)) 
cbar_kws = {'orientation':"horizontal",
           'shrink':1,
           'extend':'min',
           'extendfrac':0.1,
           "ticks":np.arange(0,22),
           'drawedges':True,
           }
sns.heatmap(globalwarming_df, cbar_kws=cbar_kws)
```

### How to set each cell of the seaborn heatmap in a square format? 

```python
plt.figure(figsize=(16,9)) 
sns.heatmap(globalwarming_df, linewidths=3, square= True)
```

### Change x-axis labels or hide using sns.heatmap() xticklabels 

```python
plt.figure(figsize=(16,9)) 
sns.heatmap(globalwarming_df, xticklabels = np.arange(0,15))
```

### Change y-axis labels or hide using sns.heatmap() yticklabels 

```python
plt.figure(figsize=(16,9)) 
country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 
'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']
sns.heatmap(globalwarming_df, yticklabels = country_code)
```

### Heatmap without xticklabels, yticklabels and color bar – meaningless heatmap 

```python
plt.figure(figsize=(10,8)) 
sns.heatmap(globalwarming_df, xticklabels = False, yticklabels = False, cbar= False, linecolor ="w", linewidths = 1)
```

### Seaborn heatmap subplots – create multiple heatmaps 

```python
plt.figure(figsize=(30,10)) 

plt.subplot(1,3,1) # first heatmap 
sns.heatmap(globalwarming_df, cbar= False, linecolor ="w", linewidths = 1)

plt.subplot(1,3,2) # Second heatmap 
sns.heatmap(globalwarming_df, cbar= False, linecolor ="k", linewidths = 1)

plt.subplot(1,3,3) # Third heatmap 
sns.heatmap(globalwarming_df, cbar= False, linecolor ="y", linewidths = 1)
```

### Set seaborn heatmap title, x-axis, y-axis label, font size with ax (Axes) parameter 

```python
plt.figure(figsize=(16,9)) 
ax = sns.heatmap(globalwarming_df)

ax.set(title = "Heatmap",
      xlabel = "Years",
      ylabel = "Country Name",)
sns.set(font_scale = 2) # set Fontsize 2
```
