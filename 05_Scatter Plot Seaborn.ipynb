## Seaborn Scatterplot
    A scatterplot in Seaborn is used to visualize the relationship between two numerical variables.Each point represents a data observation, making it easy to identify patterns, trends, and outliers.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_df = sns.load_dataset("titanic")
titanic_df
```

### 1. Draw Seaborn Scatter Plot to Find relation between age and fare

```python
titanic_df.columns
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df)
```

### 2. hue parameter 

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex")
```

### 3. hue_order parameter 

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = "sex", hue_order = ['female', 'male'])
```

### 4. sizes parameter 

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, size = 'who', sizes = (50,300))
```

### 5. size_order parameter 

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, size = 'who', size_order = ['child', 'man', 'woman'])
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df,hue = 'who', size = 'who', size_order = ['child', 'man', 'woman'])
```

### 6. style parameter 

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, style = 'who') 
```

### 7. style-order parameter 

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, style = 'who', style_order = ['child','woman', 'man']) 

```

### 8. paletter parameter 

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = 'sex', palette = "hot") 
```

### 9. marker parameter 

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, style = 'who', markers = ['3', '1', '3']) 
```

### 10. alpha parameter 

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, alpha = 0.4) 
```

### 11. legend parameter 

```python

sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = 'sex', legend = False) 
```

### 12. axis parameter 

```python
ax = sns.scatterplot(x = "age", y = "fare", data = titanic_df,) 
ax.set( xlabel = 'age', ylabel = 'fare', title = 'Scatter Plot of Age And Fare')
```

### 13. kwargs parameter 

```python
plt.figure( figsize = (16,9))
kwargs = {'edgecolor' : 'm', 'facecolor':'k', 'linewidth':2.7, 'linestyle':'--'}
sns.scatterplot(x = "age", y = "fare", data = titanic_df, size = 'sex', sizes = (500,1000), alpha = 0.7, **kwargs) 
```

```python
sns.scatterplot(x = "age", y = "fare", data = titanic_df, hue = 'sex', palette = "magma", size = 'who', sizes = (50,300))
plt.title("Scatter Plot of Age and Fare", fontsize = 25)
plt.xlabel("Age", fontsize = 20)
plt.ylabel("Fare", fontsize = 20)
plt.savefig("Scatter Plot of Age and Fare")
plt.show()
```

```python
sns.scatterplot(x = "who", y = "fare", data = titanic_df, hue = 'alive', palette = "viridis", size = 'who', sizes = (200,500))
plt.title("Scatter Plot of Age and Fare", fontsize = 25)
plt.xlabel("Age", fontsize = 20)
plt.ylabel("Fare", fontsize = 20)
plt.savefig("Scatter Plot of Age and Fare")
plt.show()
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.scatterplot(x = "tip", y = "total_bill", data = tips_df, hue = 'sex', palette = "hot", size = 'day', sizes = (50,300), alpha = 0.7)
plt.title("Scatter Plot of Tip and Total Bill", fontsize = 25)
plt.xlabel("Tip", fontsize = 20)
plt.ylabel("Total Bill", fontsize = 20)
plt.savefig("Scatter Plot of Tip and Total Bill")
plt.show()
```

```python
plt.figure( figsize =(16,9))
kwargs = {'edgecolor':'w', 'linewidth':2, 'linestyle':'--'}
sns.scatterplot(x = "tip", y = "total_bill", data = tips_df, hue = 'sex', palette = 'Dark2', size = 'day', sizes = (200,500), **kwargs)
plt.title("Scatter Plot of Tip and Total Bill", fontsize = 25)
plt.xlabel("Tip", fontsize = 20)
plt.ylabel("Total Bill", fontsize = 20)
plt.savefig("Scatter Plot of Tip and Total Bill")
plt.show()
```
